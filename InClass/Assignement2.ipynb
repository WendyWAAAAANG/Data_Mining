```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('A2_Data.csv')
data
```

```python
Alice = [162, 60]
```

```python
height = data['Height'].tolist()
weight = data['Weight'].tolist()
```

```python
counter = 0
EdisList = []
MdisList = []
for i in zip(height, weight):
    Edis = ((i[0] - Alice[0])*(i[0] - Alice[0]) + (i[1] - Alice[1])*(i[1] - Alice[1]))**0.5
    Mdis = abs(i[0] - Alice[0]) + abs(i[1] - Alice[1])
    EdisList.append(Edis)
    MdisList.append(Mdis)
    # MdisList[counter] = Mdis
    # print(str(Edis) + ' ' + str(Mdis))
    # print('M_D = ' + str(Mdis))
    # print()
    counter += 1
```

```python
# Add data into dataframe.
data['Edis'] = EdisList
data['Mdis'] = MdisList
data
```

```python
# normalization.
h_max = data['Height'].max()
h_min = data['Height'].min()
h_r = h_max - h_min
w_max = data['Weight'].max()
w_min = data['Weight'].min()
w_r = w_max - w_min
print(h_r, w_r)
print(h_min, w_min)
print(h_max, h_min)
```

```python
h_NList = []
w_NList = []
for i in zip(height, weight):
    h_NList.append((i[0] - h_min)/h_r)
    w_NList.append((i[1] - w_min)/w_r)

data['h_N'] = h_NList
data['w_N'] = w_NList
data
```

```python
A_N = []
A_N.append((Alice[0] - h_min) / h_r)
A_N.append((Alice[1] - w_min) / w_r)
print(A_N)
```

```python
nor_height = data['h_N'].tolist()
nor_weight = data['w_N'].tolist()

counter = 0
EdisList_N = []
MdisList_N = []
for i in zip(nor_height, nor_weight):
    Edis_N = ((i[0] - A_N[0])*(i[0] - A_N[0]) + (i[1] - A_N[1])*(i[1] - A_N[1]))**0.5
    Mdis_N = abs(i[0] - A_N[0]) + abs(i[1] - A_N[1])
    EdisList_N.append(Edis_N)
    MdisList_N.append(Mdis_N)
    # MdisList[counter] = Mdis
    # print(str(Edis) + ' ' + str(Mdis))
    # print('M_D = ' + str(Mdis))
    # print()
    counter += 1
```

```python
# Add data into dataframe.
data['Edis_N'] = EdisList_N
data['Mdis_N'] = MdisList_N
data
```

```python
# use KNN which k = 5, so we select first 5 smallest elements.
data['Edis'].nsmallest(5, keep = 'all')
```
